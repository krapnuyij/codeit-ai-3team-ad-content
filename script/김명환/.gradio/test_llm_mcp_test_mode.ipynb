```python
async def wait_for_server_ready(max_wait_sec: int = 300, check_interval: int = 10) -> bool:
    """
    AI ì„œë²„ê°€ idle ìƒíƒœê°€ ë  ë•Œê¹Œì§€ ëŒ€ê¸°
    
    Args:
        max_wait_sec: ìµœëŒ€ ëŒ€ê¸° ì‹œê°„ (ì´ˆ, ê¸°ë³¸ê°’ 300ì´ˆ = 5ë¶„)
        check_interval: ìƒíƒœ í™•ì¸ ê°„ê²© (ì´ˆ, ê¸°ë³¸ê°’ 10ì´ˆ)
    
    Returns:
        True: ì„œë²„ê°€ idle ìƒíƒœ
        False: íƒ€ì„ì•„ì›ƒ ë°œìƒ
    """
    logger.info(f"ğŸ• AI ì„œë²„ idle ìƒíƒœ ëŒ€ê¸° ì¤‘... (ìµœëŒ€ {max_wait_sec}ì´ˆ)")
    
    async with MCPClient(
        base_url=MCP_SERVER_URL,
        timeout=30
    ) as client:
        elapsed = 0
        while elapsed < max_wait_sec:
            health = await client.call_tool("check_server_health", {})
            
            # healthê°€ ë¬¸ìì—´ì¸ ê²½ìš° íŒŒì‹±
            if isinstance(health, str):
                # "ìƒíƒœ: busy" ë˜ëŠ” "ìƒíƒœ: idle" íŒ¨í„´ ì°¾ê¸°
                if "idle" in health.lower():
                    logger.info(f"   ì„œë²„ idle ìƒíƒœ í™•ì¸ (ëŒ€ê¸° ì‹œê°„: {elapsed}ì´ˆ)")
                    return True
                elif "busy" in health.lower():
                    logger.info(f"   â³ ì„œë²„ busy ìƒíƒœ... {check_interval}ì´ˆ í›„ ì¬í™•ì¸ (ê²½ê³¼: {elapsed}ì´ˆ/{max_wait_sec}ì´ˆ)")
                else:
                    # ìƒíƒœë¥¼ ì•Œ ìˆ˜ ì—†ìœ¼ë©´ ë°”ë¡œ ì§„í–‰
                    logger.warning(f"   ì„œë²„ ìƒíƒœ ë¶ˆëª…: {health}")
                    return True
            else:
                # dict í˜•íƒœì¸ ê²½ìš°
                status = health.get("status", "unknown") if isinstance(health, dict) else "unknown"
                if status == "idle":
                    logger.info(f"   ì„œë²„ idle ìƒíƒœ í™•ì¸ (ëŒ€ê¸° ì‹œê°„: {elapsed}ì´ˆ)")
                    return True
                elif status == "busy":
                    logger.info(f"   â³ ì„œë²„ busy ìƒíƒœ... {check_interval}ì´ˆ í›„ ì¬í™•ì¸ (ê²½ê³¼: {elapsed}ì´ˆ/{max_wait_sec}ì´ˆ)")
                else:
                    logger.warning(f"   ì„œë²„ ìƒíƒœ: {status}")
                    return True
                
            await asyncio.sleep(check_interval)
            elapsed += check_interval
        
        logger.warning(f"   â° íƒ€ì„ì•„ì›ƒ: {max_wait_sec}ì´ˆ ë™ì•ˆ ì„œë²„ê°€ idle ìƒíƒœê°€ ë˜ì§€ ì•ŠìŒ")
        return False

logger.info("ì„œë²„ ëŒ€ê¸° í—¬í¼ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ")
```

## 7. ë¹„ë™ê¸° ì‘ì—… íŒ¨í„´: Job ID ê¸°ë°˜ í´ë§

ëŒ€ìš©ëŸ‰ ì‘ì—…ì˜ ê²½ìš° `wait_for_completion=False`ë¡œ ì„¤ì •í•˜ì—¬ ì¦‰ì‹œ job_idë¥¼ ë°›ê³ , ìƒíƒœë¥¼ ì£¼ê¸°ì ìœ¼ë¡œ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# Step 1: ê´‘ê³  ìƒì„± ì‘ì—… ì‹œì‘ (job_id ë°˜í™˜)
async def start_ad_generation():
    """
    ë¹„ë™ê¸° ì‘ì—… ì‹œì‘: Job IDë¥¼ ë°›ì•„ì„œ ë°˜í™˜
    """
    logger.info("=" * 60)
    logger.info("Step 1: ê´‘ê³  ìƒì„± ì‘ì—… ì‹œì‘")
    logger.info("=" * 60)
    
    async with MCPClient(
        base_url=MCP_SERVER_URL,
        timeout=600
    ) as client:
        
        # í…ŒìŠ¤íŠ¸ë¥¼ ìœ„í•´ ëª¨ë“  ì‘ì—… ì¤‘ë‹¨
        logger.info("ê¸°ì¡´ ì‘ì—… ì¤‘ë‹¨ ì‹œë„ ì¤‘...")
        try:
            cancel_result = await client.call_tool("cancel_all_generations", {})
            logger.info(f"ì‘ì—… ì¤‘ë‹¨ ê²°ê³¼: {cancel_result}")
        except Exception as e:
            logger.warning(f"ì‘ì—… ì¤‘ë‹¨ ì¤‘ ì˜¤ë¥˜ (ë¬´ì‹œí•˜ê³  ê³„ì†): {e}")
        
        # ì ì‹œ ëŒ€ê¸°
        await asyncio.sleep(2)
        
        params = {
            "product_image_path": str(product_image_path),
            "background_prompt": "Vibrant traditional Korean market scene, colorful fruit stalls, warm lighting",
            "text_content": "ë§›ìˆëŠ”ë°”ë‚˜ë‚˜ 2500ì› ì‹œì¥ ìíŒ",
            "text_prompt": "Bold 3D Korean text with yellow-gold gradient",
            "font_name": "NanumGothicBold.ttf",
            "composition_mode": "overlay",
            "text_position": "center",
            "test_mode": True,
            "wait_for_completion": False,  # ì¦‰ì‹œ job_id ë°˜í™˜
            "save_output_path": str(banana_output_path)
        }
        
        logger.info("ê´‘ê³  ìƒì„± ìš”ì²­ ì „ì†¡ ì¤‘...")
        start_result = await client.call_tool("generate_ad_image", params)
        logger.info(f"Job ì‹œì‘ë¨: {start_result}")
        
        # job_id ì¶”ì¶œ
        if isinstance(start_result, dict):
            job_id = start_result.get("job_id")
        elif isinstance(start_result, str):
            # ë¬¸ìì—´ ì‘ë‹µì—ì„œ job_id ì¶”ì¶œ
            import re
            match = re.search(r'ì‘ì—… ID:\s*([a-f0-9\-]+)', start_result)
            if match:
                job_id = match.group(1)
            else:
                job_id = None
        else:
            job_id = None
        
        if not job_id:
            logger.warning("job_idë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ë™ê¸° ë°©ì‹ìœ¼ë¡œ ì²˜ë¦¬ëœ ê²ƒ ê°™ìŠµë‹ˆë‹¤.")
            return start_result
        
        logger.info(f"   Job ID: {job_id}")
        logger.info("   ë‹¤ìŒ ì…€ì—ì„œ ì´ job_idë¡œ ìƒíƒœë¥¼ í™•ì¸í•˜ì„¸ìš”.")
        
        return job_id
        
# ì‹¤í–‰
job_id = await start_ad_generation()
```

```python
# Step 2: ì‘ì—… ìƒíƒœ ì¡°íšŒ (í´ë§) - ì™„ë£Œ ì‹œ ì´ë¯¸ì§€ ì €ì¥ í¬í•¨
async def check_ad_generation_status(
    job_id: str, 
    save_result_path: str = None,
    max_attempts: int = 30, 
    interval: int = 10
):
    """
    Job IDë¡œ ì‘ì—… ìƒíƒœë¥¼ í™•ì¸í•˜ê³  ì™„ë£Œ ì‹œ ì´ë¯¸ì§€ë¥¼ ì €ì¥
    
    Args:
        job_id: í™•ì¸í•  ì‘ì—… ID
        save_result_path: ì™„ë£Œ ì‹œ ì €ì¥í•  ì´ë¯¸ì§€ ê²½ë¡œ (ì¤‘ìš”!)
        max_attempts: ìµœëŒ€ ì‹œë„ íšŸìˆ˜ (ê¸°ë³¸ê°’: 30íšŒ)
        interval: í™•ì¸ ê°„ê²© (ì´ˆ, ê¸°ë³¸ê°’: 10ì´ˆ)
    
    Returns:
        ìµœì¢… ìƒíƒœ ê²°ê³¼
    """
    logger.info("=" * 60)
    logger.info(f"Step 2: ì‘ì—… ìƒíƒœ í™•ì¸ (Job ID: {job_id})")
    if save_result_path:
        logger.info(f"ì™„ë£Œ ì‹œ ì €ì¥ ê²½ë¡œ: {save_result_path}")
    logger.info("=" * 60)
    
    async with MCPClient(
        base_url=MCP_SERVER_URL,
        timeout=30
    ) as client:
        
        attempt = 0
        while attempt < max_attempts:
            await asyncio.sleep(interval)
            attempt += 1
            
            # ìƒíƒœ í™•ì¸ (save_result_path ì „ë‹¬!)
            status_params = {"job_id": job_id}
            if save_result_path:
                status_params["save_result_path"] = save_result_path
            
            status_result = await client.call_tool(
                "check_generation_status",
                status_params
            )
            
            logger.info(f"   [{attempt}/{max_attempts}] ìƒíƒœ: {status_result}")
            
            # ì™„ë£Œ ì—¬ë¶€ í™•ì¸
            if isinstance(status_result, dict):
                status = status_result.get("status")
                progress = status_result.get("progress", 0)
                
                if status == "completed":
                    logger.info("âœ… ì‘ì—… ì™„ë£Œ!")
                    if save_result_path:
                        logger.info(f"   ì´ë¯¸ì§€ ì €ì¥ë¨: {save_result_path}")
                    return status_result
                elif status == "failed":
                    logger.error(f"âŒ ì‘ì—… ì‹¤íŒ¨: {status_result.get('error')}")
                    return status_result
                else:
                    logger.info(f"   ì§„í–‰ ì¤‘... (ì§„í–‰ë¥ : {progress}%)")
            else:
                # ë¬¸ìì—´ ì‘ë‹µ ì²˜ë¦¬
                if "completed" in str(status_result).lower():
                    logger.info("âœ… ì‘ì—… ì™„ë£Œ!")
                    if save_result_path:
                        logger.info(f"   ì´ë¯¸ì§€ ì €ì¥ë¨: {save_result_path}")
                    return status_result
                elif "failed" in str(status_result).lower():
                    logger.error(f"âŒ ì‘ì—… ì‹¤íŒ¨: {status_result}")
                    return status_result
        
        logger.warning(f"â° íƒ€ì„ì•„ì›ƒ: {max_attempts * interval}ì´ˆ ë™ì•ˆ ì‘ì—…ì´ ì™„ë£Œë˜ì§€ ì•ŠìŒ")
        return {"status": "timeout", "message": "ì‘ì—…ì´ ì œí•œ ì‹œê°„ ë‚´ì— ì™„ë£Œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤."}
        

# ì‹¤í–‰ (job_idë¥¼ ì• ì…€ì—ì„œ ë°›ì•„ì•¼ í•¨)
if 'job_id' in dir() and job_id:
    status_result = await check_ad_generation_status(
        job_id, 
        save_result_path=str(banana_output_path)  # ì €ì¥ ê²½ë¡œ ì „ë‹¬!
    )
else:
    logger.warning("job_idê°€ ì—†ìŠµë‹ˆë‹¤. ë¨¼ì € ì´ì „ ì…€ì„ ì‹¤í–‰í•˜ì„¸ìš”.")
```
